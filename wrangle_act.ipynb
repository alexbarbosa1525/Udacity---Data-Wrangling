```python
tweets_list = []
errors = []

for tweet_id in twitter_links['tweet_id']:
    try:
        page = api.get_status(tweet_id, tweet_mode = 'extended')
        curtidas = page['favorite_count']
        retweets = page['retweet_count']
        data = page['created_at']
        
        tweets_list.append({'tweet_id': int(tweet_id),
                        'curtidas': int(curtidas),
                        'retweets': int(retweets),
                        'data': pd.to_datetime(data)})
    
    except Exception as e:
        print(str(tweet_id)+ " _ " + str(e))
        errors.append(tweet_id)
```
