```python
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import datetime as dt
import seaborn as sns
```

```python
twitter_insights = pd.read_csv('twitter_archive_master.csv')
```

```python
twitter_insights.info()
```

```python
twitter_insights
```

1 - Vamos começar vendo quais as clasificações mais comuns entre os cachorros. Temos as quatro normais (doggo, puppo, pupper, floofer) e a classificação especial (doggo e pupper) que é quase toda formada por pais e filhotes.

```python

dog_class = twitter_insights.groupby('dog_classification').count()
dog_class_count = dog_class.tweet_id
```

```python
dog_class_count
```

```python
dog_class_count.plot(kind ='barh', align='center', alpha=0.5)
plt.xlabel('Average Dog Rating')
plt.title('Number of photos')
plt.show()
```

Vemos que temos muito mais pupper's que todos as demais classificações.

Agora vamos ver em relação a média de curtidas para cada classificação.

```python
anl_curtidas = twitter_insights[twitter_insights['dog_classification'] != 'None'].groupby('dog_classification')['curtidas'].mean()
```

```python
anl_curtidas
```

```python
anl_curtidas.plot(kind ='barh', align='center', alpha=0.5)
plt.xlabel('Average Dog Rating')
plt.title('Likes per class')
plt.show()
```

Agora temos um dado interessante.
Apesar de ter muito mais fotos, a classificação pupper tem a menor média.

Poderiamos pensar que a média diminui devido a grande quantidade de valores.
Mas se as fotos mantivessem um determinado nivel de curtidas, a média seria maior.


```python
anl_curtidas_soma = twitter_insights[twitter_insights['dog_classification'] != 'None'].groupby('dog_classification')['curtidas'].sum()
```

```python
anl_curtidas_soma
```

Confirmado. Apesar de ter apenas 63 fotos com contagem de curtidas no dataset, a classificação doggo tem 1.273.770 curtidas, e a pupper tem 1.433.801 curtidas em 203 fotos.
Uma diferença considerável.

Vamos ver se há correlação com o numero de retweets.

```python
anl_retweets = twitter_insights[twitter_insights['dog_classification'] != 'None'].groupby('dog_classification')['retweets'].mean()
```

```python
anl_retweets
```

```python
anl_retweets.plot(kind ='barh', align='center', alpha=0.5)
plt.xlabel('Average Dog Rating')
plt.title('Retweet Rate')
plt.show()
```

Novamente. Pupper tem a menor média, e a classificação doggo tem a maior média.

Agora vamos analisar a relação da avaliação recebida com o numero de retweets e curtidas.

```python
anl_avaliacao = twitter_insights.groupby(['Avaliação'])['retweets','curtidas'].mean()
```
